### Seaborn in for Data Analytics


```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
## styling 


# "darkgrid"  dark background + grid lines
# "whitegrid"  white background + grid lines
# "white"  plain white
# "dark"  plain dark
# "ticks"  white background + ticks visible

sns.set_style("white")

#set context

# "notebook" (default)  normal size
# "paper"  chhota, reports ke liye
# "talk"  thoda bada
# "poster"  presentation ke liye bahut bad


#palette
# "Set1"
# "Set2"
# "Set3"
# "Pastel1"
# "Pastel2"
# "Dark2"
# "Accent"
# "tab10"
# "tab20"

sns.set_context("notebook")
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", hue = "sex",palette="Dark2")
sns.despine(left=True)
```

```python
sns.stripplot(data=penguins,x="species",y="body_mass_g",hue="island",dodge=True)
```

```python
sns.swarmplot(data=penguins,x="species",y="body_mass_g",hue="island",dodge=True)
```

```python
sns.histplot(data=penguins, x="body_mass_g", bins=20,hue="sex",multiple="stack")
```

```python
sns.regplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    ci=95,
    order=2,
    color="Green"
)
```

```python
sns.set_style("whitegrid")
sns.lineplot(
    data=penguins,
    x="body_mass_g",
    y="flipper_length_mm",
    color="Green",
    hue="sex",
    # style="sex"
)
```

```python
#joint plot   

 #  y hist
    #    
    #    
    #      scatter / regression
    #    
    #     x hist



sns.jointplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    kind="scatter",
    hue="sex"
)
```

```python
sns.barplot(data=penguins, x="species", y="body_mass_g",estimator=sum,hue="sex")
```

```python
sns.countplot(data=penguins, x="species")
```

```python
# || < Maximum (whisker)
# |    Outlier  |
# | |
# |            |
# |  Q3   Q1   |
# | |
# |Median|
# || < Minimum (whisker)



sns.boxplot(data=penguins, x="species", y="body_mass_g",hue="sex")
```

```python
sns.violinplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    hue="sex"
)
```

```python
sns.kdeplot(
    data=penguins,
    x="body_mass_g",
    hue="species",
   palette="Dark2",
   fill="True"
)
```

```python
penguins.head(2)
columns=["bill_length_mm",	"bill_depth_mm"	,"flipper_length_mm"	,"body_mass_g"]
sns.heatmap(data=penguins[columns].corr(),annot=True,cmap="Reds",linewidths="2",linecolor="black")
```

```python
sns.heatmap(penguins.isnull(), cbar=False)
```

```python
sns.rugplot(data=penguins, x="body_mass_g",height=0.2)
```

```python
sns.pairplot(data=penguins, hue="species")
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.add_legend()
```

#  **Seaborn EDA  Konsa Plot Kab Use Hotta Hai?**

---

#  **1. Scatter Plot**

 **Use when:**

* Do **two numeric variables** ka *relationship / pattern* dekhna hai.
* Example: Flipper length vs Body mass.

 **Why use:**

* Correlation
* Trend
* Clusters
* Outliers

---

#  **2. Strip Plot**

 **Use when:**

* **Categorical (x-axis)** vs **Numeric (y-axis)**
* Har individual point dekhna hai.

 **Useful for:**

* Distribution
* Category-wise spread
* Overlaps

---

#  **3. Swarm Plot**

 **Use when:**

* Strip plot jaisa, par **points overlap nahi karte**  gap dekar arrange hote hain.
* Clean, perfect category distribution.

 **Better than strip plot** when data thoda zyada ho.

---

#  **4. Histogram**

 **Use when:**

* **Single numeric variable ka distribution** dekhna hai.

 **What it shows:**

* Shape (Normal? Skewed?)
* Range
* Frequency
* Gaps
* Outliers

 **Hue:**

* Category wise histogram (male vs female etc.)

---

#  **5. KDE Plot**

 **Use when:**

* Histogram ka **smooth version** chahiye.
* Single numeric variable ka clean distribution dekhna hai.

 **What KDE shows:**

* Peaks
* Skewness
* Bimodal distribution
* Density

 **Use hue:**
Category-wise distribution  smooth curves.

---

#  **6. Regression Plot (regplot)**

 **Use when:**

* Two numeric variables me **linear / polynomial trend** dekhna hai.

 **Shows:**

* Regression line
* CI (confidence interval)
* Trend (positive / negative)

---

#  **7. Line Plot**

 **Use when:**

* **Time-series** ya **continuous x-variable**.
* Trend dekhna hai.

 **Examples:**

* Sales growth
* Sensor data
* Stock price

---

#  **8. Joint Plot**

 **Use when:**

* Aapko **ek hi frame me 3 cheezein** dekhni hain:

 Scatter plot (relationship)
 X histogram
 Y histogram

 **Great for:**

* Correlation + distribution combined

---

#  **9. Bar Plot**

 **Use when:**

* Category-wise **mean / sum / count** show karna hai.

 Example:

* Species  avg body mass
* City  total sales

---

#  **10. Count Plot**

 **Use when:**

* Category ki **frequency (count)** dekhni hai.

 Example:

* Kitne Adelie, Kitne Gentoo species
* Kitne male vs female

---

#  **11. Box Plot**

 **Use when:**

* Numeric distribution + outliers clearly dekhne hai.

 **Shows:**

* Median
* Quartiles (Q1, Q3)
* IQR
* Outliers
* Whiskers
* Shape idea (skewness via median position)

 Best for category-wise comparisons.

---

#  **12. Violin Plot**

 **Use when:**

* Distribution + density + shape dekhna ho
* Boxplot + KDE combo

 **Shows:**

* Skewness
* Peaks
* Spread
* Bimodality

 **Better than boxplot** when data shape important ho.

---

#  **13. Heatmap**

 **Use when:**

* Correlation matrix visualize karna ho
* Missing values check karna ho
* Matrix-like data show karna ho

 Example:
`sns.heatmap(df.corr(), annot=True)`

---

#  **14. Rug Plot**

 **Use when:**

* Numeric variable ka **exact raw data position** dikhana ho.
* KDE + Rug best combo.

 Looks like: | | | | | lines on axis.

---

#  **15. Pair Plot**

 **Use when:**

* Dataset ka **full EDA overview** chahiye.
* Sare numeric columns ke relationships + distributions.

 **Shows:**

* Diagonal  histogram / KDE
* Upper/lower  scatter, kde
* Hue  category separation

---

#  **16. PairGrid**

 **Use when:**

* Pairplot ka **custom version** chahiye.
* Har section pe custom chart map karna hai.

Example:

* diagonal  hist
* upper  scatter
* lower  kde

---

#  **Summary Table (Cheat Sheet)**

| Plot      | Use                                     |
| --------- | --------------------------------------- |
| Scatter   | Relationship (numeric vs numeric)       |
| Strip     | Category vs numeric (raw points)        |
| Swarm     | Clean non-overlapping strip             |
| Histogram | Single variable distribution            |
| KDE       | Smooth distribution curve               |
| Regplot   | Trend + regression line                 |
| Lineplot  | Time-series trend                       |
| Jointplot | Relationship + individual distributions |
| Barplot   | Mean/sum of categories                  |
| Countplot | Category counts                         |
| Boxplot   | Distribution + outliers quickly         |
| Violin    | Distribution shape + density            |
| Heatmap   | Correlation / missing data              |
| Rugplot   | Raw distribution ticks                  |
| Pairplot  | All columns EDA overview                |
| PairGrid  | Fully customized pairplot               |
